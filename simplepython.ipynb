```python
cereals_mean = cereals['Sugars'].transform('mean')
```

```python
cereals_std = cereals['Sugars'].transform('std')
```

```python
cereal = (
    cereal
    .assign(
        cereals_mean = cereals['Sugars'].transform('mean'),
        cereals_std = cereals['Sugars'].transform('std')


    )
)
```

```python
cereal['Normalized_Sugars'] = (cereal['Sugars']-cereal['cereals_mean'])/cereal['cereals_std']
```
